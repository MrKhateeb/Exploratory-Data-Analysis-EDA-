Q6 Which category sells more and in which state ?!

```python
ss.columns
```

```python
category_state_sales = ss.groupby(['Product Category', 'State or Province'])['Sales'].sum().reset_index()
total_sales_by_category = category_state_sales.groupby('Product Category')['Sales'].sum().reset_index()
total_sales_by_category
```

```python
top_category = total_sales_by_category.loc[total_sales_by_category['Sales'].idxmax()]
top_category
```

```python
top_category_states = category_state_sales[category_state_sales['Product Category'] == top_category['Product Category']]
top_state = top_category_states.loc[top_category_states['Sales'].idxmax()]
top_state
```

```python
# Print results
print(f"The category with the highest total sales is: {top_category['Product Category']}")
print(f"The state with the highest sales for this category is: {top_state['State or Province']}")
```

```python
category_state_sales = ss.pivot_table(index='State or Province', columns='Product Category', values='Sales', aggfunc='sum')
category_state_sales.plot(kind='bar', stacked=True, figsize=(14,8), colormap='tab20')
plt.title('Category Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.show()
```

Q17 What is the most effective delivery  way ?!

```python
ss['Ship Mode'].value_counts()
```

```python
category_hist = sns.FacetGrid(ss, col='Ship Mode')
category_hist.map(plt.hist, 'Product Category')
plt.show()
```

```python
shipModeGR = ss.groupby('Ship Mode')
avg_deli_time = shipModeGR['Day of delivey'].mean()
avg_deli_time.reset_index()
```

```python
most_effective_deliv = avg_deli_time.idxmin()  # Ship Mode with the shortest average delivery time
most_effective_deliv
```

```python
shortest_time = avg_deli_time.min()
shortest_time
```

```python
print(f"The most effective delivery way is: {most_effective_deliv} with an average delivery time of {shortest_time} days.")
```
